# Load thư viện

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from abc import abstractmethod
```

# 1: Implement lại thuật toán Linear Regression

```python
class LinearRegression():
    def __init__(self,X,Y,N):
        # Để có thể làm tốt các bài sau, các bạn cần có kiến thức nền tảng về Objects Oriented Programming (OOP)
        # Mình khuyến khích bài post này:

        # Giải thích đơn giản về OOP
        # https://www.freecodecamp.org/news/object-oriented-programming-concepts-21bb035f7260/

        # OOP trong Python
        # https://realpython.com/python3-object-oriented-programming/
        # https://www.geeksforgeeks.org/abstract-classes-in-python/
        self.x = X
        self.y = Y
        self.n = N
    def w(self):
      return self.w

    def fit(self):
        # Hàm fit sẽ là nơi chứa logic của thuật toán Linear Regression
        # Các bạn hay xoá "..." ở phía dưới đây, và implement code
        self.new_x = np.hstack((np.ones((self.n,1)),self.x))
        self.w = np.dot(np.linalg.inv(np.dot(self.new_x.T,self.new_x)),np.dot(self.new_x.T,self.y))


    def predict(self):
        # Như tên của hàm, đây là nơi để các bạn dự đoán dữ liệu
        # Như hướng dẫn phía trên, hãy xoá "..." ở phía dưới đây,
        # và implement code dự đoán.
        return np.dot(self.new_x,self.w).reshape(1,-1)[0]
```

# 2a: Fit dữ liệu trong file linear.csv và visualize

```python
# Visualize data
data = pd.read_csv("/content/drive/MyDrive/Machine_Learning/LAB 1_T6/LAB 1.1/linear.csv").values
x = data[:,0].reshape(-1,1)
y = data[:,1].reshape(-1,1)
n = data.shape[0]
# fit data
regressor = LinearRegression(x,y,n)
regressor.fit()
predict = regressor.predict()
plt.plot(x,predict,'r')
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
# Code trong cell này.
```

# 2b: Tính loss Root Mean Squared Error - RMSE

```python
# Code trong cell này.
w = regressor.w
del regressor
diff = x * w[1] + w[0]
error = diff - y
squared_error = np.power(error,2)
#squared_error
mean_squared_error = np.sum(squared_error)/n
#mean_squared_error
root_mean_squared_error = np.sqrt(mean_squared_error)
print(root_mean_squared_error)
```

# 4: Cải thiện model

```python
# Code trong cell này.
data = pd.read_csv("/content/drive/MyDrive/Machine_Learning/LAB 1_T6/LAB 1.1/nonLinear.csv").values
x = data[:,0].reshape(-1,1)
new_x = np.hstack((x.copy(),x**2))
y = data[:,1].reshape(-1,1)
n = data.shape[0]

regressor = LinearRegression(new_x,y,n)
regressor.fit()
pre = regressor.predict()
plt.plot(x,pre,'r')
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()


# Calculate RMSE (Root Mean Square Error)
output = y.reshape(1,-1)[0]
error = pre - output
squared_error = np.power(error,2)
mean_squared_error = np.sum(squared_error)/n
root_mean_squared_error = np.sqrt(mean_squared_error)
print(root_mean_squared_error)
```
